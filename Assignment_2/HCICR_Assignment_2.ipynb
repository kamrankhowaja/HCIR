#### Explain briefly the ethno-socio linguistic model for communication in humans.

The ethno-Socio linguistic model for communication in human is based beyond sender-reciver models, it clarifies that communication context, where the meaning of a message can change depending on several parameters ;ike Temporal(When), Spatial(Where), Socio-culture(With whom) and external factors such as presence of ojects and audiences also influences. each individual has its personal set of goals to express and interpret communication signs. The core of this model is whether intentional or unintentional, we humans are always contineously communicating like even being silence in a particular context reflects communication. The model also talks about meta communication, which means communication about communication is also seen between conversation where temporarily orignal conversation is stop to seek more clarification on the communcation happened 

#### Name the components that constitute spoken dialog systems. What is the role of each of these components?

ASR - Automatic Speech Recognition - this analyses acoustic signal from user and recognize the spoken words.

SLU - Spoken Language Understanding - This analyses the input form ASR to extract intents(querry, greeting, instruct etc) and entities(name , place , thing, time) 

DM - Dialog Management - The important part composed of Dialog Context Model(DCM), responsible for tracking the dialog state and information from NLU/SLU and Dialog Policy Model(DPL), responsible for making decision for next dialog act(intent,entities) considering the information given in DCM.

NLG - Natural Language Generation - it converst the dialog act into a sequence of Words

TTS - Text to Speech - it simply converts the words into the acoustic signals which are actually the response.
